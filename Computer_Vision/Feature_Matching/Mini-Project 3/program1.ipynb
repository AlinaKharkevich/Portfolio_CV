```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('photo_1.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = np.float32(gray)

corners = cv.goodFeaturesToTrack(gray, maxCorners = 4, qualityLevel=0.1, minDistance=1, useHarrisDetector=True, k=0.04)
corners = np.int0(corners)
print(corners[0].ravel())

for i in corners:
    x,y = i.ravel()
    cv.circle(img, (x,y), 10, 255, -1)

plt.figure(figsize = (10, 10))
plt.imshow(img), plt.axis('off')
plt.show()
#cv.imwrite("1.1_Harrison_4_corners_img.jpg", img)
```

```python
img = cv.imread('photo_1.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp, des = sift.detectAndCompute(gray, None)

#img = cv.drawKeypoints(gray, kp, img)
img = cv.drawKeypoints(gray, kp, img, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize = (10, 10))
plt.imshow(img), plt.axis('off')
```
